### Working with large datasets and reading in chunks

```python
for df in pd.read_csv(r"C:\Users\Abdulkadir\Documents\Programming\Personal\DataScience\pokemon_data.csv", chunksize=10):
    print(f'Chunk')
    print(df)
```

```python
# empty df with columns from chunks size df
new_df = pd.DataFrame(columns=df.columns)
new_df
```

```python
for df in pd.read_csv(r"C:\Users\Abdulkadir\Documents\Programming\Personal\DataScience\pokemon_data.csv", chunksize=10):
    results = df.groupby(['Type 1']).count()
    new_df = pd.concat([new_df, results])

new_df[['#']]
```
