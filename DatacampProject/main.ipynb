```python
# rename columsn
renamed_cols = {
    'client_id': 'id',
    'campaign': 'number_contacts',
    'previous': 'previous_campaign_contacts',
    'poutcome': 'previous_outcome',
    'y': 'campaign_outcome',
    'euribor3m': 'euribor_three_months',
    'nr_employed': 'number_employed'
}

df.rename(columns=renamed_cols, inplace=True)
```

```python
df
```

```python
# create client table
client = pd.DataFrame(columns=[
    'id', 
    'age', 
    'job', 
    'marital', 
    'education', 
    'credit_default',
    'housing', 
    'loan'
])

# create campaign table
campaign = pd.DataFrame(columns=[
    'campaign_id',
    'client_id',
    'number_contacts',
    'contact_duration',
    'pdays',
    'previous_campaign_contacts',
    'previous_outcome',
    'campaign_outcome',
    'last_contact_date'
])

# create economics table
economics = pd.DataFrame(columns=[
    'client_id',
    'emp_var_rate',
    'cons_price_idx',
    'euribor_three_months',
    'number_employed'
])
```

```python
# add data into client table
client[[
    'id', 
    'age', 
    'job', 
    'marital', 
    'education', 
    'credit_default',
    'housing', 
    'loan'
]] = \
df[[
    'id', 
    'age', 
    'job', 
    'marital', 
    'education', 
    'credit_default',
    'housing', 
    'loan'
]]

# add data into campaign table
campaign[[
    'client_id',
    'number_contacts',
    'contact_duration',
    'pdays',
    'previous_campaign_contacts',
    'previous_outcome',
    'campaign_outcome',
]] = \
df[[
    'id',
    'number_contacts',
    'duration',
    'pdays',
    'previous_campaign_contacts',
    'previous_outcome',
    'campaign_outcome',
]]

# add data into economics
economics[[
    'client_id',
    'emp_var_rate',
    'cons_price_idx',
    'euribor_three_months',
    'number_employed'
]] = \
df[[
    'id',
    'emp_var_rate',
    'cons_price_idx',
    'euribor_three_months',
    'number_employed'
]]
```

```python
# change . to _ in education column
client['education'] = client['education'].apply(lambda x: x.replace('.', '_'))
client['education'].unique()
```

```python
# change unknown value to NULL
client = client.apply(lambda x: x.replace('unknown', np.nan), axis=1)
campaign = campaign.apply(lambda x: x.replace('unknown', np.nan), axis=1)
economics = economics.apply(lambda x: x.replace('unknown', np.nan), axis=1)
```

```python
# Convert "success" and "failure" in the "previous_outcome" and "campaign_outcome" columns 
# to binary (1 or 0) and nonexistent to nan
campaign['previous_outcome'] = campaign['previous_outcome'].map({'success':1, 'failure':0, 'nonexistent': np.nan})
```
