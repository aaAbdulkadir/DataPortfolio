```python
import snscrape.modules.twitter as sntwitter
import pandas as pd

query = "$AAPL"
tweets = []
limit = 10

# stock list
list_of_stocks = pd.read_html('https://en.wikipedia.org/wiki/Nasdaq-100')[4]['Ticker'].to_list()
stock_list = []

# add cashtag to filter for stocks
for stock in list_of_stocks:
    stock_list.append(f'${stock}')

for stock in stock_list:
    for tweet in sntwitter.TwitterSearchScraper(stock).get_items():
        if len(tweets) == limit:
            break
        else:
            tweets.append([tweet.date, tweet.username, tweet.content])
    print(f'{stock} completed')
            
df = pd.DataFrame(tweets, columns=['Date', 'User', 'Tweet'])
```
